```python
df['Product'].value_counts().plot(kind='bar')
plt.show()
```

 The bar chart indicates that printers recorded the highest number of sales among all products. Tables had the second-highest sales, followed by chairs. Phones recorded the lowest number of sales, making them the least sold product in the dataset. This suggests that printers were the most popular product during the period under study, while phones experienced the lowest demand.

```python
df['ReferralSource'].value_counts().plot(kind='bar')
plt.show()
```

The chart shows that Instagram is the leading referral source, generating the highest number of referrals. Email ranks second in terms of referrals, while Referral is the least significant source, contributing the lowest number of referrals. This indicates that Instagram is the most effective channel for driving traffic or customer engagement, whereas the Referral channel has the lowest impact.

```python
yearl_sales = df.groupby('Date')['TotalPrice'].sum().reset_index()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

monthly_sales = df.groupby(df['Date'].dt.month)['TotalPrice'].sum().reset_index()

sns.lineplot(data=monthly_sales, x='Date', y='TotalPrice', marker='o')

plt.show()
```

## KEY OBSERVATIONS

The analysis shows that Instagram is the most effective referral source, generating the highest customer referrals, while the Referral channel contributes the least, indicating Instagram’s strong role in driving traffic and engagement.

In terms of product performance, printers recorded the highest sales, making them the most popular product. Tables and chairs also performed well, while phones recorded the lowest sales, suggesting lower customer demand.

Monthly revenue fluctuated significantly across the year, peaking in June and reaching its lowest in September. Overall, the first half of the year outperformed the second half, with steady growth up to June followed by a decline and mild fluctuations, suggesting seasonal effects in customer behavior.

Outlier analysis shows no significant outliers, indicating a consistent and reliable dataset. Overall, Instagram drives most customer acquisition, printers are the top-performing product, and business performance was strongest in the first half of the year, especially in June.
